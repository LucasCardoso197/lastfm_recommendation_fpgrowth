```python
import pyspark
sc = pyspark.SparkContext(appName="Music_Recomendation")
```

```python
from pyspark.sql import SparkSession

spark = SparkSession \
    .builder \
    .appName("CorrelationExample") \
    .getOrCreate()
```

```python
datasetPath = "./lastfm-dataset-360K/usersha1-artmbid-artname-plays.tsv"
df1 = spark.read.csv(datasetPath, sep=r'\t', header=False).select('_c0','_c2')
```

```python
import pyspark.sql.functions as F
finalDf = df1.groupBy('_c0').agg(F.collect_set('_c2').alias('items')).withColumnRenamed("_c0", "user")
print((finalDf.count(), len(finalDf.columns)))
```

```python
from pyspark.ml.fpm import FPGrowth, FPGrowthModel
loadedModel = FPGrowthModel.load("mymodel")

# predict
predictionsDF = loadedModel.transform(finalDf).show(truncate=False)
```

```python
df = spark.createDataFrame([
    (0, ['radiohead', 'muse', 'coldplay']),
    (1, ['iron maiden', 'megadeth']),
    (2, ['snoop dogg', 'eminem'])
], ["id", "items"])
loadedModel.transform(df).show(truncate=False)
```
